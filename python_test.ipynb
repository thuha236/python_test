Hàm dùng list đơn thuần

```python
def f_list(list):
    uniq = []
    count = []
    for a in list:
        if a in uniq:
            index = uniq.index(a)
            count[index]+=1
        else:
            uniq.append(a)
            count.append(0)
            index= len(uniq)-1
        print(count[index],end=' ')
```

Hàm dùng dictionary

```python
def f_dict(list):
    dict={}
    for a in list:
        if a in dict.keys():
            dict[a]+=1
        else:
            dict[a]=0
        print(dict[a],end=' ')
```

test 1: list =  [9,8,7,9,8,7,1,2,1,1,2,3]

```python
list =  [9,8,7,9,8,7,1,2,1,1,2,3]
f_list(list)
print()
f_dict(list)
```

test 2: with a list of 100,000 random number

```python
#create list of 100,000 random numbers between 1 and 100
import random
list=[]
for i in range (100000):
    x = random.randint(1,100)
    list.append(x)
    
#####test
import time

#test function using list
start1 = time.time()
f_list(list)
end1 = time.time()


#test function using dictionary
start2 = time.time()
f_dict(list)
end2 = time.time()

print()
print('Execution time of function using list: ', end1-start1)
print('Execution time of function using dict: ', end2-start2)

```
